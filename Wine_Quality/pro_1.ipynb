```python
columns_to_plot = ['alcohol', 'fixed acidity', 'volatile acidity', 'citric acid', 
                   'residual sugar', 'chlorides', 'free sulfur dioxide', 
                   'total sulfur dioxide', 'density', 'pH', 'sulphates']

# Add empty slots to fill the 12 subplots
columns_to_plot += [None] * (12 - len(columns_to_plot))

# Create subplots
fig, axes = plt.subplots(nrows=3, ncols=4, figsize=(18, 12))

# Plot each column's distribution
for i, column in enumerate(columns_to_plot):
    row, col = divmod(i, 4)
    if column is not None and column in red_df.columns:
        sns.histplot(red_df[column], bins=30, kde=True, edgecolor='black', ax=axes[row, col], color='blue')  # Adjust color as needed
        axes[row, col].set_title(f'Overall Distribution of {column.capitalize()}')
        axes[row, col].set_xlabel(f'{column.capitalize()} Level')
        axes[row, col].set_ylabel('Count')
    else:
        axes[row, col].axis('off')
        axes[row, col].set_title(f'No Data for {column.capitalize() if column else "Empty"}')

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Filter the data for quality 3
quality_3_data = red_df[red_df['quality'] == 3]

# Melt the DataFrame to long format for FacetGrid
melted_data = quality_3_data.melt(value_vars=['alcohol', 'fixed acidity', 'volatile acidity', 
                                                'citric acid', 'residual sugar', 'chlorides', 
                                                'free sulfur dioxide', 'total sulfur dioxide', 
                                                'density', 'pH', 'sulphates'], 
                                    var_name='Variable', value_name='Value')

# Create a FacetGrid to show histograms for each variable
g = sns.FacetGrid(melted_data, col='Variable', col_wrap=3, height=4)
g.map(sns.histplot, 'Value', bins=30, kde=True, edgecolor='black')

# Add titles and labels
g.set_titles(col_template='{col_name}')
g.set_axis_labels('Value', 'Count')

plt.tight_layout()
plt.show()
```
