Columns:
- `year` : The manufacturing year of the vehicle.
- `make` : The brand or manufacturer of the vehicle.
- `model`: The specific model of the vehicle.
- `trim` : The additional designation for the vehicle model.
- `body` : The body type of the vehicle (e.g., SUV, Sedan).
- `transmission` : The type of transmission in the vehicle (e.g., automatic).
- `vin` : Vehicle Identification Number, a unique code for each vehicle.
- `state` : The state where the vehicle is registered.
- `condition` : Condition of the vehicle, possibly rated on a scale.
- `odometer` : The mileage or distance traveled by the vehicle.
- `color` : The exterior color of the vehicle.
- `interior` : The interior color of the vehicle.
- `seller` : The entity selling the vehicle.
- `mmr` : Manheim Market Report, possibly indicating the estimated market value of the vehicle.
- `sellingprice` : The price at which the vehicle was sold.
- `saledate` : The date and time when the vehicle was sold.

## Business Analysis

- **1. Finance KPIs**: focus on revenue, profitability, and cost efficiency.
    - **Total Sales**
        - **Formula**: `Sum of sellingprice`
        - **Description**: Total revenue generated from vehicle sales.
        - **Plot**: Bar chart showing total sales over time (by month/quarter).
    - **Average Selling Price**
        - **Formula**: `Total Sales / Number of Transactions`
        - **Description**: The average price at which vehicles are sold, useful for identifying market trends and pricing strategies.
        - **Plot**: Line chart showing trends in average selling price over time.
    - **Price to MMR Ratio**
        - **Formula**: `sellingprice / mmr`
        - **Description**: Compares the selling price to the market value (MMR). A higher ratio indicates that the vehicle was sold above the market value.
        - **Plot**: Box plot comparing selling price to MMR values, grouped by vehicle type.
    - **Profitability by Vehicle**
        - **Formula**: `Profit = sellingprice - mmr`
        - **Description**: Calculate the profitability of each sale, useful if acquisition cost data is available.
        - **Plot**: Bar chart or scatter plot showing profitability per vehicle model or seller.
---

- **2. Sales KPIs**: focus on the performance of sales transactions and sales growth.
    - **Sales by Make/Model**
        - **Formula**: `Group by make or model and sum the sellingprice`
        - **Description**: Identifies which vehicle brands or models generate the most revenue.
        - **Plot**: Bar chart comparing sales performance across different vehicle makes or models.
    - **Sales by Vehicle Condition**
        - **Formula**: `Group by condition and sum the sellingprice`
        - **Description**: Analyze how vehicle condition (rating from 1 to 5, for example) impacts the sales price.
        - **Plot**: Stacked bar chart showing sales volume based on vehicle condition.
    - **Top-Selling Vehicles**
        - **Formula**: `Group by model and sum the sellingprice`
        - **Description**: Identifies which vehicle models are the most popular based on sales.
        - **Plot**: Horizontal bar chart displaying top-selling models.
    - **Sales by Body Type/Transmission Type**
        - **Formula**: `Group by body or transmission and sum the sellingprice`
        - **Description**: Helps understand the demand for specific vehicle body types (e.g., SUV, Sedan) or transmission types (e.g., automatic, manual).
        - **Plot**: Side-by-side bar chart comparing sales by body type and transmission.
    - **Sales by State**
        - **Formula**: `Group by state and sum the sellingprice`
        - **Description**: Tracks regional vehicle sales performance.
        - **Plot**: Heatmap or bar chart showing sales distribution by region.

---

- **3. Market Behaviour KPIs**: give insights into market behavior and trends over time.
    - **Price Fluctuations Over Time**
        - **Formula**: `Track sellingprice by month/quarter using saledate`, Track `sellingprice` by month or quarter using `saledate`.
        - **Description**: Identifies pricing trends over time, useful for understanding market cycles.
        - **Plot**: Line chart showing vehicle price trends over time.
    - **Vehicle Age Impact on Price**
        - **Formula**: `Vehicle Age = Current Year - year`
        - **Description**: Analyze how the age of a vehicle impacts its selling price.
        - **Plot**: Scatter plot showing the relationship between vehicle age and selling price.
    - **Odometer Impact on Price**
        - **Formula**: `Analyze relationship between odometer and sellingprice`
        - **Description**: Understand how mileage affects the vehicle's resale value.
        - **Plot**: Scatter plot or bubble chart visualizing odometer reading vs. selling price.
    - **Market Share by Make/Model**
        - **Formula**: `Group by make or model and calculate percentage of total sales`
        - **Description**: Identifies the market share distribution of  vehicle brands or models.
        - **Plot**: Pie chart showing market share distribution of different vehicle makes.

---

- **4. Operations KPIs**: focus on operational efficiency and vehicle conditions.
    - **Average Condition Score**
        - **Formula**: `Sum of condition / Number of Vehicles`
        - **Description**: Measures the overall quality of the vehicles sold.
        - **Plot**: Line chart showing average condition scores over time or by seller.
    - **Time to Sale**
        - **Formula**: `Time to Sale = saledate - Date of Vehicle Listing`
        - **Description**: Measures how quickly vehicles are sold after being listed. Useful for inventory turnover analysis.
        - **Plot**: Histogram or bar chart showing the distribution of time-to-sale for different vehicle categories.
    - **Seller Performance**
        - **Formula**: `Group by seller and sum the sellingprice`
        - **Description**: Analyzes which sellers generate the most sales, useful for partnerships and supply chain analysis.
        - **Plot**: Bar chart or heatmap showing sales performance by seller.
    - **Inventory Turnover**
        - **Formula**: `Number of Vehicles Sold / Average Inventory`
        - **Plot**: Line chart showing inventory turnover trends by vehicle category or seller.
